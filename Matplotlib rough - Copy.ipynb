```python
#BAR GRAPH
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use('seaborn')
width=0.25
Ages_x=(16,17,18,19,20,21,22,23)
X=np.arange(len(Ages_x))
Names=['Sai','Akshith','Manish','Raks','Koushik','Anirudh','Ram','Gattu']
plt.bar(X-width,Names,width=width,label='Names',color='#EE1289')
Height=[5.10,5.6,5.6,5.7,5.8,5.5,5.11,6.0]
plt.bar(X,Height,width=width,label='Height',color='#444444')
Weight=[8,5,6,2,5.5,7,7.4,7.7]
plt.bar(X+width,Weight,width=width,label='weight',color='#A2CD5A')
plt.legend()
plt.ylabel(' Names Height Weight')
plt.xlabel('Ages')
plt.xticks(ticks=X,labels=Ages_x)
plt.title('Description')
plt.tight_layout()
plt.show()
```

```python
# PIE PLOT
import matplotlib.pyplot as plt

Percentage=[30,80,50,60,90]
Labels=['Ram','Raks','Rio','Remo','Ravi']
explode=[0,0,0.6,0,0]
plt.pie(Percentage,labels=Labels,explode=explode,wedgeprops={'edgecolor':'white'},shadow=True,startangle=180,autopct='%1.1f%%')
plt.show()
```

```python
# STACK PLOT
import matplotlib.pyplot as plt

Min=[1,2,3,4,5,6,7,8,9]
player1=[1,1,2,3,3,4,4,5,2]
player2=[2,1,2,1,1,1,4,5,3]
player3=[1,1,1,2,2,2,3,5,4]
colors=['#7CFC00','blue','#00C78C']
labels=['player1','player2','player3']
plt.stackplot(Min,player1,player2,player3,labels=labels,colors=colors)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
#FILLING AREA ON LINE PLOTS
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use('seaborn')

Ages_x=(16,17,18,19,20,21,22,23)
#X=np.arange(len(Ages_x))
Names=['Sai','Akshith','Manish','Raks','Koushik','Anirudh','Ram','Gattu']
#plt.plot(Ages_x,Names,color='#EE1289')
Height=[5.1,5.5,5.9,7,8,10,3.9,6.0]
plt.plot(Ages_x,Height,label='Height',color='#444444')
Weight=[8,5,6,2,5.5,7,7.4,7.7]
plt.plot(Ages_x,Weight,label='weight',color='#A2CD5A')
plt.legend()
overall_median=5.6
plt.fill_between(Ages_x,Weight,where=[Height<=Weight],interpolate=True,alpha=0.25)
plt.ylabel(' Names Height Weight')
plt.xlabel('Ages')

plt.title('Description')
plt.tight_layout()
plt.show()
```

```python
#PLOTTING LIVE DATA IN REAL TIME
import random
from itertools import count
import pandas as pd
%matplotlib  notebook
import matplotlib.pyplot as plt
%matplotlib inline




from matplotlib.animation import FuncAnimation
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    plt.cla()
    plt.plot(x_vals, y_vals)
    
ani=FuncAnimation(plt.gcf(),animate,interval=1000)

plt.tight_layout()
plt.show()
```

```python
#SUB PLOTS
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.gcf()
plt.gca()

Ages_x=(16,17,18,19,20,21,22,23)

Names=['Sai','Akshith','Manish','Raks','Koushik','Anirudh','Ram','Gattu']
plt.subplot(3,1,1)
plt.plot(Ages_x,Names,label='Names',color='#EE1289',linestyle="--")
plt.ylabel(' Names ')
plt.xlabel('Ages')
plt.title('Description')

Height=[5.10,5.6,5.6,5.7,5.8,5.5,5.11,6.0]
plt.subplot(3,1,2)
plt.plot(Ages_x,Height,label='Height',color='#444444',linestyle=":")
plt.ylabel(' Height')
plt.xlabel('Ages')
plt.title('Description')

Weight=[8,5,6,2,5.5,7,7.4,7.7]
plt.subplot(3,1,3)
plt.plot(Ages_x,Weight,label='weight',color='#A2CD5A')

plt.legend()
plt.ylabel('  Weight')
plt.xlabel('Ages')
plt.title('Description')
plt.tight_layout()
plt.show()
```
